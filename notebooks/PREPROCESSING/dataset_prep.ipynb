```python
def save_frequency_bands(rem_df, nrem_df, wake_df, save_path):
    delta_rem_df = rem_df[(rem_df.Frequency >= 1) & (rem_df.Frequency < 5)]
    delta_freq = ['Delta']*len(delta_rem_df)
    delta_rem_df['Frequency_Band'] = delta_freq
    theta_rem_df = rem_df[(rem_df.Frequency >= 5) & (rem_df.Frequency < 10)]
    theta_freq = ['Theta']*len(theta_rem_df)
    theta_rem_df['Frequency_Band'] = theta_freq
    sigma_rem_df = rem_df[(rem_df.Frequency >= 10) & (rem_df.Frequency < 16)]
    sigma_freq = ['Sigma']*len(sigma_rem_df)
    sigma_rem_df['Frequency_Band'] = sigma_freq
    beta_rem_df = rem_df[(rem_df.Frequency >= 16) & (rem_df.Frequency < 30)]
    beta_freq = ['Beta']*len(beta_rem_df)
    beta_rem_df['Frequency_Band'] = beta_freq
    gamma_rem_df = rem_df[(rem_df.Frequency >= 30) & (rem_df.Frequency < 48)]
    gamma_freq = ['Gamma']*len(gamma_rem_df)
    gamma_rem_df['Frequency_Band'] = gamma_freq
    high_gamma_rem_df = rem_df[(rem_df.Frequency >= 52) & (rem_df.Frequency <= 100)]
    high_gamma_freq = ['High Gamma']*len(high_gamma_rem_df)
    high_gamma_rem_df['Frequency_Band'] = high_gamma_freq
    
    delta_nrem_df = nrem_df[(nrem_df.Frequency >= 1) & (nrem_df.Frequency < 5)]
    delta_freq_nrem = ['Delta']*len(delta_nrem_df)
    delta_nrem_df['Frequency_Band'] = delta_freq_nrem
    theta_nrem_df = nrem_df[(nrem_df.Frequency >= 5) & (nrem_df.Frequency < 10)]
    theta_freq_nrem = ['Theta']*len(theta_nrem_df)
    theta_nrem_df['Frequency_Band'] = theta_freq_nrem
    sigma_nrem_df = nrem_df[(nrem_df.Frequency >= 10) & (nrem_df.Frequency < 16)]
    sigma_freq_nrem = ['Sigma']*len(sigma_nrem_df)
    sigma_nrem_df['Frequency_Band'] = sigma_freq_nrem
    beta_nrem_df = nrem_df[(nrem_df.Frequency >= 16) & (nrem_df.Frequency < 30)]
    beta_freq_nrem = ['Beta']*len(beta_nrem_df)
    beta_nrem_df['Frequency_Band'] = beta_freq_nrem
    gamma_nrem_df = nrem_df[(nrem_df.Frequency >= 30) & (nrem_df.Frequency < 48)]
    gamma_freq_nrem = ['Gamma']*len(gamma_nrem_df)
    gamma_nrem_df['Frequency_Band'] = gamma_freq_nrem
    high_gamma_nrem_df = nrem_df[(nrem_df.Frequency >= 52) & (nrem_df.Frequency <= 100)]
    high_gamma_freq = ['High Gamma']*len(high_gamma_nrem_df)
    high_gamma_nrem_df['Frequency_Band'] = high_gamma_freq
    
    
    delta_wake_df = wake_df[(wake_df.Frequency >= 1) & (wake_df.Frequency < 5)]
    delta_freq_wake = ['Delta']*len(delta_wake_df)
    delta_wake_df['Frequency_Band'] = delta_freq_wake
    theta_wake_df = wake_df[(wake_df.Frequency >= 5) & (wake_df.Frequency < 10)]
    theta_freq_wake = ['Theta']*len(theta_wake_df)
    theta_wake_df['Frequency_Band'] = theta_freq_wake
    sigma_wake_df = wake_df[(wake_df.Frequency >= 10) & (wake_df.Frequency < 16)]
    sigma_freq_wake = ['Sigma']*len(sigma_wake_df)
    sigma_wake_df['Frequency_Band'] = sigma_freq_wake
    beta_wake_df = wake_df[(wake_df.Frequency >= 16) & (wake_df.Frequency < 30)]
    beta_freq_wake = ['Beta']*len(beta_wake_df)
    beta_wake_df['Frequency_Band'] = beta_freq_wake
    gamma_wake_df = wake_df[(wake_df.Frequency >= 30) & (wake_df.Frequency < 48)]
    gamma_freq_wake = ['Gamma']*len(gamma_wake_df)
    gamma_wake_df['Frequency_Band'] = gamma_freq_wake
    high_gamma_wake_df = wake_df[(wake_df.Frequency >= 52) & (wake_df.Frequency <= 100)]
    high_gamma_freq = ['High Gamma']*len(high_gamma_wake_df)
    high_gamma_wake_df['Frequency_Band'] = high_gamma_freq
    
    
    
    rem_concat = pd.concat([delta_rem_df, theta_rem_df, sigma_rem_df, beta_rem_df, gamma_rem_df, high_gamma_rem_df])
    nrem_concat = pd.concat([delta_nrem_df, theta_nrem_df, sigma_nrem_df, beta_nrem_df, gamma_nrem_df, high_gamma_nrem_df])
    wake_concat = pd.concat([delta_wake_df, theta_wake_df, sigma_wake_df, beta_wake_df, gamma_wake_df, high_gamma_wake_df])
    
    
    os.chdir(save_path)
    
    delta_rem_df.to_csv('delta_rem.csv')
    theta_rem_df.to_csv('theta_rem.csv')
    sigma_rem_df.to_csv('sigma_rem.csv')
    beta_rem_df.to_csv('beta_rem.csv')
    gamma_rem_df.to_csv('gamma_rem.csv')
    high_gamma_rem_df.to_csv('high_gamma_rem.csv')
    
    delta_nrem_df.to_csv('delta_nrem.csv')
    theta_nrem_df.to_csv('theta_nrem.csv')
    sigma_nrem_df.to_csv('sigma_nrem.csv')
    beta_nrem_df.to_csv('beta_nrem.csv')
    gamma_nrem_df.to_csv('gamma_nrem.csv')
    high_gamma_nrem_df.to_csv('high_gamma_nrem.csv')
    
    delta_wake_df.to_csv('delta_wake.csv')
    theta_wake_df.to_csv('theta_wake.csv')
    sigma_wake_df.to_csv('sigma_wake.csv')
    beta_wake_df.to_csv('beta_wake.csv')
    gamma_wake_df.to_csv('gamma_wake.csv')
    high_gamma_wake_df.to_csv('high_gamma_wake.csv')
    
    return rem_concat, nrem_concat, wake_concat
    
```

```python
df_ls = []
band = ['Delta', 'Theta', 'Sigma', 'Beta', 'Gamma', 'High Gamma']

for frequenc_band in band:
    for anim in som_animals:
        df = som_rem.loc[som_rem['Animal_ID'] == str(anim)]
        freq_df = df.loc[df['Frequency_Band'] == frequenc_band]
        avg_freq_power = freq_df['Power'].mean()
        brainstate = freq_df['Brainstate'].values[:1]
        genotype = freq_df['Genotype'].values[:1]
        average_freq_dict = {'Animal_ID': [anim], 'Brainstate': brainstate, 'Avg_Power': [avg_freq_power],
                          'Frequency': [frequenc_band], 'Genotype': genotype}
        try:
            average_freq_df = pd.DataFrame(data = average_freq_dict)
            df_ls.append(average_freq_df)
        except:
            print(anim)
            print(frequenc_band)
            print(avg_freq_power)
```
